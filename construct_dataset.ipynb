```python
import numpy as np
import string
```

```python
wikipedia_path = '/home/eramamur/10707-project/wikipedia_sentences_tokenised.txt'
glove_path = '/home/eramamur/10707-project/glove.6B.50d.txt'
```

```python
vocab = ["MASK"]
with open(glove_path, 'r') as f:
    for line in f:
        word = line.strip().split(" ", 1)[0]
        vocab.append(word)
vocab.append("UNK")       
print len(vocab)
```

```python
word2id = dict()
id2word = dict()
for i, word in enumerate(vocab):
    word2id[word] = i
    id2word[i] = word
```

```python
def constructNGrams(document, nGramSize=41, stepSize=320):
    X = []
    Y = []
    tokens = document.strip().split()
    tokenIndexes = []
    for token in tokens:
        curToken = token.lower()
        if not curToken in word2id:
            curToken = "UNK"
        tokenIndexes.append(word2id[curToken])
            
    for i in xrange(0, len(tokenIndexes)-nGramSize+1, stepSize):
        start = i
        end = i+nGramSize
        nGram = tokenIndexes[start:end]
        if not word2id["UNK"] in nGram:
            X.append(nGram[0:nGramSize/2]+[word2id["MASK"]]+nGram[nGramSize/2+1:len(nGram)])
            Y.append(nGram[nGramSize/2])
    return X, Y
```

```python
totalSamples = 0
X=None
Y=None
with open(wikipedia_path, 'r') as f:
    curDoc=""
    docCount = 0
    for i,line in enumerate(f):
        if line.startswith("<doc id="):
            docCount+=1
            if docCount%100==0:
                print docCount, "documents processed, ", totalSamples, "samples obtained"
            curDoc=""
        elif line.startswith("</doc>"):
            newNgrams, newY = constructNGrams(curDoc, nGramSize=41)
            if not len(newNgrams) == 0:
                if X is None:
                    X = newNgrams
                    Y = newY
                else:
                    for val in newNgrams:
                        X.append(val)
                    for val in newY:
                        Y.append(val)
            totalSamples += len(newNgrams)
        else:
            curLine = line.strip()
            if curLine[len(curLine)-1] in string.punctuation:
                curLine = curLine[0:len(curLine)-1]+" "+curLine[len(curLine)-1]
            curDoc = " ".join([curDoc, curLine])
```

```python
X=np.array(X, dtype=np.int)
np.save('X.npy', X)
Y = np.array(Y, dtype=np.int).reshape((len(Y),))
np.save('Y.npy', Y)
```

```python
numTotal = X.shape[0]
print numTotal
```

```python
numValid = int(0.2*numTotal)
numTest = int(0.2*numTotal)
numTrain = numTotal-numTest-numValid
numHeldOut = numValid+numTest
```

```python
heldOutIndexes = np.random.choice(numTotal, size=(numHeldOut,), replace=False)
trainIndexes = set(range(numTotal)).difference(set(heldOutIndexes))
trainIndexes = np.array(list(trainIndexes), dtype=np.int).reshape((numTrain,))
```

```python
Xtrain = X[trainIndexes]
Ytrain = Y[trainIndexes]
XheldOut = X[heldOutIndexes]
YheldOut = Y[heldOutIndexes]
```

```python
validIndexes = np.random.choice(numHeldOut, size=(numValid,), replace=False)
testIndexes = set(range(numHeldOut)).difference(set(validIndexes))
testIndexes = np.array(list(testIndexes), dtype=np.int).reshape((numTest,))
```

```python
Xtest = XheldOut[testIndexes]
Ytest = YheldOut[testIndexes]
Xvalid = XheldOut[validIndexes]
Yvalid = YheldOut[validIndexes]
```

```python
#filtering out those labels from test and validation set that do not appear in training set

trainLabelSet = set(Ytrain)
testIndexes = []
for i,val in enumerate(Ytest):
    if val in trainLabelSet:
        testIndexes.append(i)
testIndexes = np.array(testIndexes).reshape((len(testIndexes),))

validIndexes = []
for i,val in enumerate(Yvalid):
    if val in trainLabelSet:
        validIndexes.append(i)
validIndexes = np.array(validIndexes).reshape((len(validIndexes),))
```

```python
newXtest = Xtest[testIndexes]
newYtest = Ytest[testIndexes]
newXvalid = Xvalid[validIndexes]
newYvalid = Yvalid[validIndexes]
```

```python
np.save('Xtrain.npy', Xtrain)
np.save('Ytrain.npy', Ytrain)
np.save('Xtest.npy', newXtest)
np.save('Ytest.npy', newYtest)
np.save('Xvalid.npy', newXvalid)
np.save('Yvalid.npy', newYvalid)
```
